```python
thislist = list(('apple', 'banana', 'cherry'))  # note the double round-brackets
print(thislist)
# access items in a list
print(thislist[0])  # first item
# negative indexing
print(thislist[-1])  # last item
# range of indexes
print(thislist[0:2])  # items from index 0 to index 2
```

```python
# extend list
thislist = ['apple', 'banana', 'cherry']
tropical = ['mango', 'pineapple', 'papaya']
thislist.extend(tropical)  # extend thislist by adding elements from tropical
print(thislist)
```

```python
# add tuple in list
thislist = ['apple', 'banana', 'cherry']
thistuple = ('kiwi', 'orange')
thislist.extend(thistuple)
print(thislist)
```

```python
# remove list item
thislist = ['apple', 'banana', 'cherry']
thislist.remove('banana')  # remove 'banana' from the list
print(thislist)
del thislist
```

```python
thislist = ['apple', 'banana', 'cherry']
thislist.pop(1)  # remove the item at index 1
del thislist[0]
print(thislist)  # prints ['apple', 'cherry']
```

```python
fruits = ['apple','banana', 'kiwi','mango', 'test']
newlist = []
for x in fruits:
    if 'a' in x:
        newlist.append(x)
print(newlist)

print("------")
newlist1 = [x for x in fruits if 'e' in x]
print(newlist1)
newlist2 = [x for x in fruits if x !='apple']
print(newlist2)
newlist3 = [x for x in fruits]
print(newlist3)
```

```python
thislist = ['orange', 'mango','kiwi','pineapple','banana']
thislist.sort() # alphabetically as first letter
print(thislist)
thislist1 = [11,22,45,100,1,2, 46,87,90]
thislist1.sort()
print(thislist1)
print("----------")
# sort in descending
thislist1.sort(reverse=True)
print(thislist1)
```

```python
# customize sort function
def myfunc(n):
    return abs(n-50)
thislist = [100, 50, 65, 82, 23]
thislist.sort(key=myfunc)
print(thislist)
```

```python
thislist = ['apple', 'Banana', 'cherry','Kiwi', 'Orange']  # uper case letter 
thislist.sort()  
print(thislist)
print("-----------")
thislist.sort(key=str.lower)
print(thislist)

print("------ ^^^^^^^^ -----")

mylist = thislist.copy()
print(mylist)
yourlist = list(thislist)
print(f'your list:{yourlist}')
```
