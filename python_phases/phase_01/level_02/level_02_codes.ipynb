#### 1. Take three sides and check if they form a valid triangle.

```python
side_1 = float(input("Enter side 1: "))
side_2 = float(input("Enter side 2: "))
side_3 = float(input("Enter side 3: "))


if side_1 + side_2 > side_3:
    if side_1 + side_3 > side_2:
        if side_2 + side_3 > side_1:
            print("Yes, valid triangle.")
        else:
            print("No, it's not valid.")
    else:
            print("No, it's not valid.")
else:
            print("No, it's not valid.")
```

#### 2. If the sides form a valid triangle, determine whether it is equilateral, isosceles, or scalene. 

```python
side1 = float(input("Enter side 1: ")); print(side1)
side2 = float(input("Enter side 2: ")); print(side2)
side3 = float(input("Enter side 3: ")); print(side3)

# check valid triangle
if (side1 + side2 > side3) and (side1 + side3 > side2) and (side2 + side3 > side1) :
    print("Yes, valid triangle.")
    
    # Equilateral Triangle: All three sides are equal. Every angle is also the same.
    if side1 == side2 == side3:
        print("This is Equilateral Triangle.")
        
    # Isosceles Triangle: Two sides are equal and the third side is different.
    elif side1 == side2 or side1 == side3 or side2 == side3:
        print("This is Isosceles Triangle.")
        
    # Scalene Triangle: All three sides are different from each other.No sides and angles are equal.
    elif side1 != side2 or side1 != side3 or side2 != side3:
         print("This is Scalene Triangle.")
    # else: 
    #     print("This is Scalene Triangle.")
        
else:
            print("No, it's not valid.")
```
