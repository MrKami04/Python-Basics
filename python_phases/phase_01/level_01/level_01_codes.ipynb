#### 1. Take a number and print whether it’s positive, negative, or zero.

```python
# using if-elif-else codition
# number = int(input("Enter the number: "))
number = 3
# logic is : number >0 or <0
if number > 0:
    print("Number is positive.")
elif number < 0:
    print("Number is negative.")
else:
    print("Number is zero")
    
# we right in conprehension as
print('Num is +ve') if number > 0 else print('Num is -ve') if number < 0  else print('Num is zero.')
```

#### 5. Check if a given year is a leap year. 

```python
# year = int(input("Enter the year: "))
year = 2004

# logic is: year%400==0 or (year%4==0 & year%100 !=0)
if (year % 400 == 0 ) or  (year % 4 == 0 and year % 100 != 0):
    print("Year is a leap year")
else:
    print("Year isn't a leap year.")
```

#### 8. Take a temperature value and print “Cold”, “Warm”, or “Hot” using range conditions. 

```python
temperature = 37

# logic is: >, <
if temperature < 25:
    print("Cold")
elif (temperature > 25) and (temperature <= 40):
    print("Warm")
else:
    print("Hot")
```

#### 10. Take a character and check whether it’s uppercase, lowercase, a digit, or a special character. 

```python
ch = input("Enter a character: ")

if ch.isupper():
    print("Upper case letter.")
elif ch.islower():
    print("Lower case letter.")
elif ch.isdigit():
    print("digit case.")
else:
    print("Special character.")
```
